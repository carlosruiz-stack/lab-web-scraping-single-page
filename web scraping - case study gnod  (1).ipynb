```python
# import libraries, pip install already done
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# send get request and retrieve content 
url = 'https://www.billboard.com/charts/hot-100/'
response = requests.get(url)
content = response.content
```

```python
# parse content 
soup = BeautifulSoup(content, 'html.parser')
```

```python
# find the container 
container = soup.find('div', class_='chart-results-list')
```

```python
container
```

```python
# initialise empty list with song and artist information
songs = []
artists = []
```

```python
chart_entries = soup.find_all("li", class_="chart-list__element")
song_count = 0

for entry in chart_entries:
    artist = entry.find("span", class_="chart-element__information__artist").text.strip()
    song = entry.find("span", class_="chart-element__information__song").text.strip()
    print("Artist:", artist)
    print("Song:", song)
    print("--------")

    song_count += 1
    if song_count == 100:
        break
```
